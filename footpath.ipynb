```python
import os
import cv2
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
```

```python
def plot_period(period_df):
    """
    Plot a single period (x,y) in the phase plane.
    """
    plt.figure(figsize=(6,6))
    plt.plot(period_df['avg_x'], period_df['avg_y'])
    plt.xlabel('x')
    plt.ylabel('y')
    plt.title('Average Period Trajectory')
    plt.axis('equal')
    plt.grid(True)
    plt.show()
```

```python
CSV_FILE_PATH = "footpath_representation.csv"
```

```python
 # 1. Load CSV
df = pd.read_csv(CSV_FILE_PATH)  # your file path here

df
```

```python
plt.figure(figsize=(10, 6))
plt.plot(df['avg_x'], df['avg_y'], marker='o', linestyle='-', color='b', markersize=8, label='Footpath')
for i, txt in enumerate(df['point_type']):
    plt.annotate(txt, (df['avg_x'][i], df['avg_y'][i]), textcoords="offset points", xytext=(0,10), ha='center')

plt.title('Footpath Representation', fontsize=14)
plt.xlabel('X Coordinate')
plt.ylabel('Y Coordinate')
plt.grid(True, linestyle='--', alpha=0.7)
plt.legend()
plt.show()
```
